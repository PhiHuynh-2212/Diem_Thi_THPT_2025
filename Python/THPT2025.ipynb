#### Map **'Mã ngoại ngữ'**

```python
df_thpt2025['Môn ngoại ngữ'] = df_thpt2025['Mã môn ngoại ngữ'].map(ma_ngoai_ngu)
```

#### Dùng **'Pivot table'** để tách riêng các môn ngoại ngữ

```python
df_pivot = df_thpt2025.pivot_table(
    index='SOBAODANH',
    columns='Môn ngoại ngữ',
    values='Ngoại ngữ'
).reset_index()
```

#### Kiểm tra lại df_pivot

```python
df_pivot
```

#### Merge lại với df_thpt2025 để lấy các điểm các môn ngoại ngữ

```python
df_thpt2025 = df_thpt2025.merge(df_pivot, on='SOBAODANH', how='left')
```

```python
df_thpt2025
```

#### Tạo thêm cột tổ hợp các môn, điều kiện cả 3 môn không bị NaN

```python
for ma_khoi, mon_list in to_hop_diem.items():
    mask = df_thpt2025[mon_list].notna().all(axis=1)

    df_thpt2025.loc[mask, ma_khoi] = df_thpt2025.loc[mask, mon_list].sum(axis=1)
```

#### Transform các cột điểm về **float**

```python
for col in df_thpt2025.columns:
    if col not in ['SOBAODANH', 'Mã môn ngoại ngữ','Môn ngoại ngữ']:      
        df_thpt2025[col] = df_thpt2025[col].astype(float)
```
